```python
# Create a scatter plot
plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Fare'], color='blue')
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=df, x="Age", y="Fare")
plt.show()
```

```python
# create seaborn plot

sns.scatterplot(data=df, x='Age', y='Fare', hue='Sex', size='Pclass')

# customize with matplotlib
plt.title('Age vs Fare by Sex and Class')
plt.xlabel('Passenger Age (years)')
plt.ylabel('Fare paid ($)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="Age", kde=True, bins=20,edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.figure(figsize=(8, 5))
sns.distplot(df["Age"], bins=20, kde=True, 
             hist_kws={"edgecolor": "black"})
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df["Fare"], shade=True)
plt.title("Density Plot of Fare")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], 
             hue='Survived', 
             diag_kind='kde')
plt.suptitle("Pairwise Relationships of Age, Fare, Pclass, and Survival", y=1.02)
plt.show()
```

```python
# Create a Regression plot

plt.figure(figsize=(8, 6))
sns.regplot(data=df, x="Age", y="Fare", scatter_kws={'alpha':0.3})
plt.title("Linear Relationship between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# Create a Box Plot

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Pclass", y="Age", hue="Survived")
plt.title("Age Distributon by Pclass and Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Create a violin plot

plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x="Pclass", y="Age", hue="Survived", split=True)
plt.title("Age Distributon by Pclass and Survival (violin Plot)")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Create a Count plot

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="Pclass", hue="Survived", palette='magma')
plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```
